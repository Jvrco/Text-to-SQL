````python
import pandas as pd

path = path_base + 'Bases_case_final.xlsx'

telemetria = pd.read_excel(path, sheet_name = 'Telemetria')

# formatar para datetime
telemetria['Data'] = pd.to_datetime(telemetria['Data']).dt.date
telemetria.head()
````

````python
chassis = pd.read_excel(path, sheet_name = 'Chassis')
chassis.head()
````

````python
#cria tabela
table_name = 'chassis'
chassi_table = Table(
    table_name,
    metadata_obj,
    Column("Chassi", Integer),
    Column("Contrato", Integer),
    Column("Cliente", Integer),
    Column("Modelo", Integer),
    extend_existing=False
)
#extend_existing = false para não duplicar a tabela caso precise rodar novamente esta celula
metadata_obj.create_all(engine)
````

````python
#transforma para sql e adiciona na tabela criada anteriormente
chassis.to_sql('chassis', con=engine, index = False, if_exists= 'append')
````

````python
stmt = select(
    chassi_table.c.Chassi,
    chassi_table.c.Contrato,
    chassi_table.c.Cliente,
    chassi_table.c.Modelo,
).select_from(chassi_table)

with engine.connect() as connection:
  results = connection.execute(stmt).fetchall()
  print(results)
````

````python
# mesmo processo para a tabela de telemetria.
table_name = 'telemetria'
telemetria_table = Table(
    table_name,
    metadata_obj,
    Column("Chassi", Integer),
    Column("UnidadeMedida", String, nullable = False),
    Column("Categoria", String, nullable = False),
    Column("Data", String, nullable = False),
    Column("Serie", String, nullable = False),
    Column("Valor", Float),
    extend_existing=False
)
metadata_obj.create_all(engine)
````

````python
telemetria.to_sql('telemetria', con=engine, index = False, if_exists= 'append')
````

````python
#visualizar a tabela
stmt = select(
    telemetria_table.c.Chassi,
    telemetria_table.c.UnidadeMedida,
    telemetria_table.c.Categoria,
    telemetria_table.c.Data,
    telemetria_table.c.Serie,
    telemetria_table.c.Valor,
).select_from(telemetria_table)

with engine.connect() as connection:
  results = connection.execute(stmt).fetchall()
  #imprimir parte dos dados
  for i in range(10):
    print(results[i])
````

````python
from llama_index.core.indices.struct_store.sql_query import (SQLTableRetrieverQueryEngine,)
from llama_index.core.objects import (
    SQLTableNodeMapping,
    ObjectIndex,
    SQLTableSchema,
)
from llama_index.core import VectorStoreIndex

table_node_mapping = SQLTableNodeMapping(sql_database)
# table schema and column schema para entender o contexto das tabelas
chassis_text1 = (
   "This table gives information about the machine chassis. You should use this table when asked informations about the contract number of the machine, number of the client that owns the machine or the model of the machine.\n"
   "Query examples to use this table:"
   "List the top 2 models of machines (and the quantity) with the most machines.\n "
   "Wich client has the most machines?\n"
   "Wich client has the machine {machine identification number}?"
   " \n The columns refer to: \n 'Chassi': machine identification number.\n 'Contrato': contract number of the machine.\n 'Cliente': number of the client that owns the machine.\n 'Modelo': model number of the machine."
)
telemetria_text1 = (
    "Description: Base referring to the data sent by the machines of certain sensors, these machines are identified by the chassis number.\n  "
    "When to use: This table will be used when the query asks for the time in each status of the motor, for the fuel comsumption, or for the time of use for each engine mode. The reference for this calculation is the machine identification number, column 'Chassi'.\n"
    "The query can specify the sub-category of the column 'Categoria', then it will need to use the column 'Serie'.\n"
    "Query examples to use this table:\n"
    "How long was the machine {chassis number} used under low load on 2024-1-17?\n "
    "What was the fuel consumption of the machine {chassis number} on the day 2024-1-18?\n"
    "How long did the machine 523093 operate in High Power Mode in 2024-1-30? \n"
    "The columns of this table refer to:\n"
    "'Chassi':  machine identification number."
    "'UnidadeMedida': Unit of the variable being sent by the machine, l stands for liters and hr stand for hours."
    "'Categoria': Type of information being sent, 'Uso do motor' refers to the time of use of the engine in each status of the series and 'Uso da configuração do Modo do Motor ' refers to the time of use of each engine mode, they can be 'Chave-ligada', 'Marcha-Lenta', 'Carga-Baixa', 'Carga-Média', 'Carga-Alta', "
    " and 'Uso do Combustível do Motor' refers to the fuel comsumption in each of the following series 'HP','P' or 'E'."
    "'Data': Information submition date."
    "'Serie': Sub-type of the information being sent, related to the category, it can be 'Chave-ligada': engine off, 'Marcha-Lenta':Engine running but unproductive, 'Carga-Baixa':Engine running but low usage, 'Carga-Média':Engine running with regular use, "
    "'Carga-Alta':Engine running with intense use, 'HP':High Power Mode ,'P': default power mode, or 'E': economy power mode."
    "'Valor': Value of the measurement being sent."
)

telemetria_text2 = ('''
telemetria table:
Description: The 'telemetria' table, records the information that evolve the engine of the machine. Database with the information obtained by sensors for each chassis. The informations includes the categorie (column 'Categoria') and the series (column 'Serie')
that provides the engine status. Along with that, the value and the unity of the measurement are present.

Columns:

'Chassi':
Description: A unique identifier of each machine. It's the machine identification number.
Usage: The machine identification number is used to gather all the informations sent by the sensors, it identifies the information detailed above.
-------------------
'UnidadeMedida':
Description: It's the phisycal unity of the measurement sent by the sensor, the measurement is given by the column 'Valor'.
Usage: This column is used to specify the physical unity of the measurement 'Valor'. The values of this column can be 'l' that stands for liters and 'hr' that stands for hours.
--------------------
'Categoria':
Description: Type of information beig sent by the sensor. It caracterizes the measurement 'Valor' present in the row.
Usage: It's used to classify the information, it can assume three values, 'Uso do Motor', 'Uso do Combustível do motor' e 'Uso da Configuração do Modo do Motor'.
Values of 'Categoria':
'Uso do Motor': time in each status of the engine. The status is given by the column 'Serie'.
'Uso do Combustível do Motor': fuel consumption of the engine in each status of the engine. The status is given by the column 'Serie'.
'Uso da Configuração do Modo do Motor': refers to the time of use of each status of the engine. The status is given by the column 'Serie'.
--------------------
'Data':
Description: The date of obtaining the engine and the machine information.
Usage: Used to obtain informations during a period of time or in specifics days. It can be used to sum the total time in one status of the engine or the total fuel consumption in determined period of time.
---------------
'Serie':
Description: Status of the 'Categoria'. It categorizes each value of the column 'Categoria'.
Usage: It categorizes each value of the column 'Categoria'. For example, the query can specify the engine status presented in the column 'Serie' to retrieve the measurement.
Values of 'Serie':
'Chave-ligada': Engine off.
'Marcha Lenta': Engine running but unproductive.
'Carga Baixa': Engine running but low usage.
'Carga Média': Engine running with regular use.
'Carga Alta': Engine running with intense use.
'HP': High Power Mode.
'P': default power mode.
'E': economy power mode.
------------
'Valor':
Description: Measurement obtained by the sensors. The values are categorized by the columns 'Chassi', 'Categoria' and 'Serie'.
Usage: It is used, mainly, to calculate sums of fuel consuptions and time in each engine status. Any sum operation in telemetry table must be in this column.


''')

chassis_text2 = ('''
chassis table:
Description: The 'chassis' table records the information, that does not involve the use of the machine, of each machine and relevante details, as the owner, the model number of the machine, the contract number of the macninhe.
Columns:

Chassi:
Description: A unique identifier of each machine. It's the machine identification number. It represents the machine.
Usage: The machine indentification number, chassis number, is used to identifie the machine, leading to the identification of the relevant details mentioned above.

Contrato:
Description: Contract number of the machine.
Usage: This number is used for contractual and legal purposes, to indentify each contract with the correspondig machine.

Cliente:
Description: number of the client that owns the machine.
Usage: This column can be used to retrieve the machines of each client, or to inform abot the contracts or the machines that the client owns.

Modelo:
Description: model number of the machine.
Usage: This column can be used to inform the model associated to each machine. You can count the number of unique machines associated to each model number.


''')

# a organização dos schemas 2 ajudam na interpretação das querys

table_node_mapping = SQLTableNodeMapping(sql_database)
table_schema_objs = [
    (SQLTableSchema(table_name='chassis', context_string=chassis_text2)),
    (SQLTableSchema(table_name='telemetria', context_string=telemetria_text2)),
]



obj_index = ObjectIndex.from_objects(
    table_schema_objs,
    table_node_mapping,
    VectorStoreIndex,
)

````

````python
from llama_index.core.retrievers import NLSQLRetriever

nl_sql_retriever = NLSQLRetriever(
    sql_database, tables=["chassis", "telemetria"], return_raw=False, llm = llm,
)
````

````python
from llama_index.core.query_engine import RetrieverQueryEngine
#retriever para sintetizar as respostas
query_engine = RetrieverQueryEngine.from_args(nl_sql_retriever, llm = llm)
````

````python
instructions = '''
You are an expert SQL data analist. Given the tables chassis or telemetria you should answer a SQL query, following the bellow instructions.
1. Your SQL lines should not have ```.
2. Your final answer should be easy to understand and well developed, as if you were a sales representant talking to a client.
3. After the answer, you should append the SQL query used.
4. You must not halucinate, use the data providen and only the columns of the tables.
5. Under any circunstance you should change or delete the data.
'''
````

````python
#query do usuario do sistema

query_user = input('Digite sua consulta (datas devem ser no formato YYYY-MM-DD): ')
````

````python
from googletrans import Translator
#traduzir a query
translator = Translator()
detected_language = translator.detect(query_user).lang

query_user_translated = translator.translate(query_user, src=detected_language, dest = 'en').text
print(query_user_translated)

# seria interessante uma parte para formatar a data para formato YYYY-MM-DD
````
